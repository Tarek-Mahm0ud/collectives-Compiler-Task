```python
%%writefile main.py
from mpi4py import MPI

comm = MPI.COMM_WORLD
rank = comm.Get_rank()

if rank == 0:
    data_to_scatter = [1, 2, 3, 4]
    multiplier = 2
else:
    data_to_scatter = None
    multiplier = None

multiplier = comm.bcast(multiplier, root=0)
my_number = comm.scatter(data_to_scatter, root=0)

my_result = my_number * multiplier
total_sum = comm.reduce(my_result, op=MPI.SUM, root=0)
new_stock = comm.gather(my_result, root=0)

if rank == 0:
    print(f"Total Cost: {total_sum}")
    print(f"New Stock: {new_stock}")
```
